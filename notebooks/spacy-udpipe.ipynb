```python
import spacy_udpipe
```

```python
spacy_udpipe.download('nl')
```

```python
text = 'Volgens het kabinet en de deskundigen van het Outbreak Management Team is de avondklok een belangrijk middel in de strijd tegen corona en met name ook tegen de derde golf die door allerlei besmettelijkere varianten van het virus dreigt. Rutte zei gisteren dat er voor dinsdag, als er weer een persconferentie over de maatregelen is gepland, nog eens kan worden gekeken of de avondklok op of voor 2 maart kan worden opgeheven. "Maar op dit moment is dat echt niet verstandig."'
print(text)
```

```python
nlp = spacy_udpipe.load('nl')
```

```python
doc = nlp(text)
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.dep_)
```

```python
text_tokenized = [[token.text for token in s] for s in doc.sents ]
```

Process further with stanza:

```python
import stanza
import e2edutch.stanza
import stroll.stanza
```

```python
snlp = stanza.Pipeline(lang='nl', processors='tokenize,lemma,pos,depparse,srl,coref',
                      tokenize_pretokenized=True)
```

```python
sdoc = snlp(text_tokenized)
```

```python
sdoc
```

```python
# Check that stanza has the exact same tokens

compare_list = []
stanza_length = 0
for sent_udpipe, sent_stanza in zip(doc.sents, sdoc.sentences):
    assert len(sent_udpipe) == len(sent_stanza.words)
    for tok_udpipe, tok_stanza in zip(sent_udpipe, sent_stanza.words):
        assert tok_udpipe.text == tok_stanza.text
        
        compare_list.append([
            ('lemma', tok_udpipe.lemma_, tok_stanza.lemma),
            ('pos', tok_udpipe.pos_, tok_stanza.upos),
            ('xpos', tok_udpipe.tag_, tok_stanza.xpos),
            ('head', tok_udpipe.head.i, stanza_length+tok_stanza.head-1),
            ('dep', tok_udpipe.dep_, tok_stanza.deprel)
        ])
    stanza_length += len(sent_stanza.words)
```

```python
# Investigate differences between the two taggers

for l in compare_list:
    for prop, a, b in l:
        if str(a) != str(b):
            print('{}: {} = {} / {}'.format(l[0][1], prop, a, b))
```
